```python
# For droping the data and filling the null values - Data cleaning data 3
data3.drop(['Other'], inplace=True)
data3['Name'] = data3['Name'].fillna('All Regions')
data3.drop(columns=['Deaths - newly reported in last 24 hours',], inplace=True)
```

```python
# data cleaning sheet 2
mean_value = data2['TOTAL_VACCINATIONS_PER100'].mean()
data2['TOTAL_VACCINATIONS_PER100'].fillna(mean_value, inplace=True)
data2.drop(data2[data2['COUNTRY'] == 'Bonaire'].index, inplace=True)
data2.drop(data2[data2['COUNTRY'] == 'Eritrea'].index, inplace=True)
data2.drop(data2[data2['COUNTRY'] == 'Sint Eustatius'].index, inplace=True)
data2.drop(data2[data2['COUNTRY'] == 'Saba'].index, inplace=True)
data2.drop(['ISO3', 'FIRST_VACCINE_DATE', 'PERSONS_BOOSTER_ADD_DOSE_PER100', 'PERSONS_BOOSTER_ADD_DOSE'], axis=1, inplace=True)
```
